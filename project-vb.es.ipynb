#### Ejercicio 15. Traza un gráfico de barras de la mediana de los precios y explica lo que observas (debes usar el subconjunto obtenido del Ejercicio 14) (★★★)

Imprima un gráfico de barras de la mediana de precios y escriba en la celda Markdown un breve análisis sobre el gráfico.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset
ds = pd.read_csv('assets/real_estate.csv', sep=';')

# Filtrar el DataFrame para incluir solo las poblaciones del cinturón sur de Madrid
cinturon_sur = ds[ds['level5'].isin(['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón'])]

# Calcular la mediana de los precios por población
medianas_precios = cinturon_sur.groupby('level5')['price'].median()

# Trazar el gráfico de barras
plt.figure(figsize=(10, 6))
medianas_precios.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Mediana de Precios por Población en el Cinturón Sur de Madrid')
plt.xlabel('Población')
plt.ylabel('Precio Mediano (€)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

El gráfico de barras muestra la **mediana de los precios** por población en el cinturón sur de Madrid. Observamos que las poblaciones de **Getafe** y **Leganés** tienen precios medianos más altos en comparación con **Fuenlabrada** y **Alcorcón**. Esto podría indicar que, a pesar de estar en la misma área geográfica, las características del mercado inmobiliario en ciertas poblaciones son más costosas, posiblemente debido a la ubicación, el tipo de viviendas disponibles o la demanda de estas zonas. La mediana nos da una buena idea del precio típico de las viviendas en cada población, ya que no se ve tan afectada por valores extremos como la media.

#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Cargar el dataset
ds = pd.read_csv('assets/real_estate.csv', sep=';')

cinturon_sur = ds[ds['level5'].isin(['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón'])]

scaler = MinMaxScaler()

fig, ax = plt.subplots(figsize=(10, 6))

for poblacion in cinturon_sur['level5'].unique():
    # Filtrar datos por población
    datos_poblacion = cinturon_sur[cinturon_sur['level5'] == poblacion]
    
    # Normalizar los precios para la población
    precios_normalizados = scaler.fit_transform(datos_poblacion[['price']])
    
    # Trazar el histograma
    ax.hist(precios_normalizados, bins=30, alpha=0.5, label=poblacion)

# Configurar el gráfico
ax.set_title('Histogramas de Precios Normalizados por Población')
ax.set_xlabel('Precio Normalizado')
ax.set_ylabel('Frecuencia')
ax.legend(title='Población')

plt.show()
```

El gráfico de barras muestra los **histogramas de los precios normalizados** de las viviendas en el cinturón sur de Madrid (Fuenlabrada, Leganés, Getafe y Alcorcón).

#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')

cinturon_sur = ds[ds['level5'].isin(['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón'])]

fig, axes = plt.subplots(2, 2, figsize=(12, 10))  # 2 filas y 2 columnas de subgráficos

for i, poblacion in enumerate(cinturon_sur['level5'].unique()):
    ax = axes[i//2, i%2]  # Obtener el subplot correspondiente (2x2)
    
    datos_poblacion = cinturon_sur[cinturon_sur['level5'] == poblacion]
    
    ax.scatter(datos_poblacion['surface'], datos_poblacion['price'], alpha=0.5)
    
    ax.set_title(f'Dispersión: {poblacion}')
    ax.set_xlabel('Superficie (m²)')
    ax.set_ylabel('Precio (€)')
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
from ipyleaflet import Map, CircleMarker
import pandas as pd

ds = pd.read_csv("assets/real_estate.csv", sep=';')

# cinturón sur
cinturon_sur = ds[ds['level5'].isin(['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón'])]

cinturon_sur = cinturon_sur.dropna(subset=['latitude', 'longitude'])

# Crear un color para cada municipio
colores = {
    'Fuenlabrada': 'blue',
    'Leganés': 'green',
    'Getafe': 'red',
    'Alcorcón': 'purple'
}

mapa = Map(center=(40.35, -3.75), zoom=11)

for _, row in cinturon_sur.iterrows():
    color = colores.get(row['level5'], 'black')  # Color por población
    marker = CircleMarker()
    marker.location = (row['latitude'], row['longitude'])
    marker.radius = 4
    marker.color = color
    marker.fill_color = color
    marker.fill_opacity = 0.6
    mapa.add_layer(marker)

# Mostrar el mapa con los marcadores
mapa
```
